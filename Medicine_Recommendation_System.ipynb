```python
# Split features and target
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']

# Encode prognosis labels
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    Y, 
    test_size=0.3, 
    random_state=20
)
X
y
```
